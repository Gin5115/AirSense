```python
import pandas as pd
data = pd.read_csv(r'D:/snu/academic/sem6/ML_Lab/Lab3/telecom_customer_churn.csv')
```

```python
data.head()
```

```python
data.replace({'Yes':1,'No':0},inplace = True)
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
data_final = data [['Gender','Age','Married','Number of Dependents','City','Number of Referrals','Tenure in Months','Offer','Phone Service','Avg Monthly Long Distance Charges','Internet Service','Avg Monthly GB Download','Premium Tech Support','Contract','Total Revenue','Customer Status']]
```

```python
data_final.columns
```

```python
data_final.head()
```

```python
data_final.isnull().sum()
```

```python
data_final = data_final.drop(columns = ['Offer'])
data_final.columns
```

```python
data_final["Avg Monthly Long Distance Charges"] = data_final["Avg Monthly Long Distance Charges"].fillna(0)
```

```python
data_final["Avg Monthly GB Download"] = data_final["Avg Monthly GB Download"].fillna(0)
```

```python
data_final["Premium Tech Support"] = data_final["Premium Tech Support"].fillna(0)
```

```python
data_final.isnull().sum()
```

```python
data_final.head()
```

```python
string_columns = data_final.select_dtypes(include='object').columns
```

```python
for column in string_columns:
    dummy_columns = pd.get_dummies(data_final[column], prefix=column,dtype = int)
    data_final = pd.concat([data_final, dummy_columns], axis=1)
    data_final = data_final.drop([column], axis=1)
```

```python
data_final.head()
```

```python
d_end = data_final.pop("Customer Status")
data_final["Customer Status"] = d_end
data_final
```

```python
x = data_final.iloc[:,:-1].values
y = data_final.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```
