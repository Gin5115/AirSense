## the **TotalCharges** column contains null values and they might not be immediately visible in the output of data.info() and isnull() because it doesn't explicitly show the null values since it is of type Object and data present in is numeric.

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
null_values_count = data.isnull().sum()

print("Number of null values in each column:")
print(null_values_count)
```

```python
data.dropna(subset=['TotalCharges'], inplace=True)
```

```python
null_values_count = data.isnull().sum()

print("Number of null values in each column:")
print(null_values_count)
```

```python
data.head()
```

```python
data.replace({'Yes': 1, 'No': 0, 'No phone service': 0, 'No internet service': 0}, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
object_columns = data.select_dtypes(include=['object']).columns

for column in object_columns:
    data = pd.get_dummies(data, columns=[column], drop_first=True, dtype=int)
```

```python
last_column = data.pop("Churn")
data["Churn"] = last_column
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

## Some machine learning algorithms are sensitive to the scale of the input features. Feature scaling ensures that all features contribute equally to the model. It standardizes or normalizes the range of independent variables or features of the dataset.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(data[['tenure', 'MonthlyCharges', 'TotalCharges']])
```

```python
data.info()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('Churn', axis=1)
Y = data['Churn']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state = 45)
```

## Import acuuracy_score

```python
from sklearn.metrics import accuracy_score
```

# 1. Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
#dt_pred = dt.predict(X_test)
#dt_accuracy = accuracy_score(Y_test, dt_pred)
#print(f"Decision Trees Accuracy: {dt_accuracy:.2f}")
dt_accuracy = dt.score(X_test, Y_test)
dt.score(X_test, Y_test)
```

# 2. Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm.fit(X_train, Y_train)
#svm_pred = svm.predict(X_test)
#svm_accuracy = accuracy_score(Y_test, svm_pred)
#print(f"SVM Accuracy: {svm_accuracy:.2f}")
svm_accuracy = svm.score(X_test, Y_test)
svm.score(X_test, Y_test)
```

# 3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100, criterion='gini',random_state=0)
rfc.fit(X_train,Y_train)
#rfc_pred = rfc.predict(X_test) 
#rfc_accuracy = accuracy_score(Y_test, rfc_pred)
#print(f"Random Forest Accuracy: {rfc_accuracy:.2f}")
rfc_accuracy = rfc.score(X_test, Y_test)
rfc.score(X_test, Y_test)
```

# 4. Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0).fit(X_train, Y_train)
#clf_pred = clf.predict(X_test) 
#clf_accuracy = accuracy_score(Y_test, clf_pred)
#print(f"Random Forest Accuracy: {clf_accuracy:.2f}")
clf_accuracy = clf.score(X_test, Y_test)
clf.score(X_test, Y_test)
```

# Accuracy Score Comparison

# randomizedsearchCV & gridsearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

## Decision Trees

```python
param = {'max_depth':[4,5,6,7,8,9,10], 
        'criterion':['gini','entropy'],
        'splitter' :['best', 'random'],
        "max_features": [1,2,3,4,5,6,7,8,9],
        "min_samples_leaf": [1,2,3,4,5,6,7,8,9],
        }
```

### RSCV Decision Trees

```python
tree = DecisionTreeClassifier(random_state=0)
tree_rscv = RandomizedSearchCV(tree, param, n_jobs=-1)
tree_rscv.fit(X_train,Y_train)
print(tree_rscv.best_params_)
```

```python
dt = DecisionTreeClassifier(criterion= 'gini', max_depth= 8, max_features= 6, min_samples_leaf= 4, splitter= 'best')
dt.fit(X_train, Y_train)
dt.score(X_test, Y_test)
```

### GSCV Decision Trees

```python
tree = DecisionTreeClassifier(random_state=0)
tree_gscv =  GridSearchCV(tree,param,cv=10, n_jobs=-1)
tree_gscv.fit(X_train,Y_train)
print(tree_gscv.best_params_)
```

```python
dt = DecisionTreeClassifier(criterion= 'entropy', max_depth= 8, max_features= 9, min_samples_leaf= 8, splitter= 'best')
dt.fit(X_train, Y_train)
dt.score(X_test, Y_test)
```
